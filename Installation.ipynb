On vide ensuite 'darknet/cfg' et 'darknet/data' de leurs fichiers uniquement (pas des dossiers et de leur contenu).

```python
import os

def vider_dossier(path):
    for fichier in os.listdir(path):
        chemin_fichier = os.path.join(path, fichier)
        if os.path.isfile(chemin_fichier):
            os.remove(chemin_fichier)
```

```python
vider_dossier('darknet/data')
vider_dossier('darknet/cfg')
```
